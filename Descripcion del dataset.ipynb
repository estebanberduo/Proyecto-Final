# Dataset: Classification of Body Postures and Movements
### Descripción:
Este dataset muestra información sobre varios sujetos de prueba que portaron acelerómetros en distintas partes del cuerpo con el fin de determinar la postura de todo su cuerpo en base a la localización de unas partes del cuerpo.

### Clasificación de las variables:
Cualitativa          
1. User                
2. Gender
3. Class

Cuantitativa
1. How tall in meters
2. Age
3. Body mass index
4. Weight
5. X1
6. Y1
7. Z1
8. X2
9. Y2
10. Z2
11. X3
12. Y3
13. Z3
14. X4
15. Y4
16. Z4

### Resumen estadístico de las variables:
1. user (text) 
2. gender (text) 
3. age (integer) 
4. how_tall_in_meters (real) 
5. weight	(int) 
6. body_mass_index	(real) 
7. x1 (type int, contains the read value of the axis 'x' of the 1st accelerometer, mounted on waist) 
8. y1 (type int, contains the read value of the axis 'y' of the 1st accelerometer, mounted on waist) 
9. z1 (type int, contains the read value of the axis 'z' of the 1st accelerometer, mounted on waist) 
10. x2 (type int, contains the read value of the axis 'x' of the 2nd accelerometer, mounted on the left thigh)	
11. y2 (type int, contains the read value of the axis 'y' of the 2nd accelerometer, mounted on the left thigh)	
12. z2 (type int, contains the read value of the axis 'z' of the 2nd accelerometer, mounted on the left thigh)	
13. x3 (type int, contains the read value of the axis 'x' of the 3rd accelerometer, mounted on the right ankle)	
14. y3 (type int, contains the read value of the axis 'y' of the 3rd accelerometer, mounted on the right ankle)	
15. z3 (type int, contains the read value of the axis 'z' of the 3rd accelerometer, mounted on the right ankle)	
16. x4 (type int, contains the read value of the axis 'x' of the 4th accelerometer, mounted on the right upper-arm)	
17. y4 (type int, contains the read value of the axis 'y' of the 4th accelerometer, mounted on the right upper-arm)	
18. z4 (type int, contains the read value of the axis 'z' of the 4th accelerometer, mounted on the right upper-arm)
19. class (text)

### Variables de interés:
1. user: Muestra el nombre del sujeto a estudiar.
2. gender: muestra el género del sujeto a estudiar.
3. age: muesta la edad del sujeto de estudio.
4. how_tall_in_meters: muestra la estatura del sujeto de estudio en metros.
5. weight: muestra el peso del sujeto de estudio.
6. body_mass_index: muestra el índice de masa corporal del sujeto de estudio.
7. x1: contiene el valor del eje x del acelerómetro montado en la cintura.
8. y1: contiene el valor del eje y del acelerómetro montado en la cintura.
9. z1: contiene el valor del eje z del acelerómetro montado en la cintura.
10. class: muestra la postura en la que se encuentra el sujeto.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

########NOTA: LAS GRAFICAS SE DEBEN CORRER POR SEPARADO  PARA OBTENER RESULTADOS ADECUADOS
data = pd.read_csv("posturas.csv")
data.head()
data.info()

#Histogramas: Muestran las frecuencias de cada variable por medio de barras
plt.hist(data.gender.dropna())
plt.hist(data.how_tall_in_meters.dropna())
bygender=data.groupby("gender")
genero=bygender["gender"].count()
sns.distplot(genero)
plt.hist(data.weight.dropna()) #en kilogramos
plt.hist(data.body_mass_index)
plt.hist(data.x1)
plt.hist(data.x2)
plt.hist(data.x3)
plt.hist(data.x4)
plt.hist(data.y1)
plt.hist(data.y2)
plt.hist(data.y3)
plt.hist(data.y4)
plt.hist(data.z1)
plt.hist(data.z2)
plt.hist(data.z3)
plt.hist(data.z4)

#Kernel density estimation plot : Metodo no parametrico de estimar el PDF (probabilidad de densidad de una función) de una variable aleatoria continua
hombres = data[(data.gender == "Man")]
mujeres = data[(data.gender == "Woman")]
sns.kdeplot(hombres.how_tall_in_meters.dropna())
sns.kdeplot(data.how_tall_in_meters.dropna())
sns.kdeplot(data.body_mass_index.dropna())
sns.kdeplot(mujeres.body_mass_index.dropna())
sns.kdeplot(data.x1)
sns.kdeplot(data.x2)
sns.kdeplot(data.x3)
sns.kdeplot(data.x4)
sns.kdeplot(data.y1)
sns.kdeplot(data.y2)
sns.kdeplot(data.y3)
sns.kdeplot(data.y4)
sns.kdeplot(data.z1)
sns.kdeplot(data.z2)
sns.kdeplot(data.z3)
sns.kdeplot(data.z4)

#Muestra la distribución de la data adicional a un boxplot.
sns.violinplot(data.how_tall_in_meters.dropna())
sns.violinplot(data.body_mass_index)
sns.boxplot(data.body_mass_index)
sns.violinplot(data.x1)
sns.violinplot(data.x2)
sns.violinplot(data.x3)
sns.violinplot(data.x4)
sns.violinplot(data.y1)
sns.violinplot(data.y2)
sns.violinplot(data.y3)
sns.violinplot(data.y4)
sns.violinplot(data.z1)
sns.violinplot(data.z2)
sns.violinplot(data.z3)
sns.violinplot(data.z4)
```
